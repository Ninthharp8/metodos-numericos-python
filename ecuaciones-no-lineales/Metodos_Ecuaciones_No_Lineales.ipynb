## M√©todo de Bisecci√≥n

### Objetivo
Encontrar una ra√≠z de una funci√≥n en un intervalo $[a, b]$ donde $f(a)f(b) < 0$, dividiendo iterativamente el intervalo a la mitad.

### Fundamento Te√≥rico
El m√©todo se basa en el teorema del valor intermedio: si $f$ es continua en $[a, b]$ y $f(a)f(b) < 0$, entonces existe al menos una ra√≠z en ese intervalo.

En cada paso, se eval√∫a el punto medio $c = (a + b)/2$ y se reemplaza uno de los extremos del intervalo seg√∫n el signo de $f(c)$.

### Observaciones
- M√©todo robusto y garantiza convergencia.
- Convergencia lenta (lineal).
- Requiere que la ra√≠z est√© encerrada en un intervalo con signos opuestos.

## M√©todo de Punto Fijo

### Objetivo
Transformar $f(x) = 0$ en $x = g(x)$ y encontrar una soluci√≥n iterativa $x_{n+1} = g(x_n)$.

### Fundamento Te√≥rico
Si $g(x)$ es continua y cumple ciertas condiciones (como ser contractiva), entonces la iteraci√≥n converge a un punto fijo que es ra√≠z de $f(x)$.

### Observaciones
- Requiere reformular la ecuaci√≥n original.
- Convergencia depende de la derivada de $g(x)$ en la ra√≠z.
- Puede divergir si $|g'(x)| \geq 1$.

## M√©todo de Newton-Raphson

### Objetivo
Encontrar una ra√≠z aproximada usando una aproximaci√≥n lineal (tangente) a la funci√≥n en un punto cercano.

### Fundamento Te√≥rico
Dado un valor inicial $x_0$, se define la sucesi√≥n:


$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$


### Observaciones
- Convergencia cuadr√°tica si la ra√≠z es simple y \(f'(x) \ne 0\).
- Requiere el c√°lculo de la derivada.
- Puede divergir si el punto inicial est√° lejos de la ra√≠z o si \(f'(x_n)\) es muy peque√±o.

## M√©todo de la Secante

### Objetivo
Aproximar la ra√≠z sin necesidad de conocer la derivada, usando dos puntos iniciales.

### Fundamento Te√≥rico
A partir de dos puntos $x_{n-1}$ y $x_n$, se define:


$
x_{n+1} = x_n - f(x_n) \cdot \frac{x_n - x_{n-1}}{f(x_n) - f(x_{n-1})}
$

### Observaciones
- No requiere derivadas.
- Convergencia superlineal.
- Puede ser inestable si los puntos iniciales no son adecuados.
